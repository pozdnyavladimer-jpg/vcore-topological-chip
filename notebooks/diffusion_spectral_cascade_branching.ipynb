```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D33_SPECTRAL_CASCADE_BRANCHING
# Branching spectral cascade across multiple receiver cells
#
# Goal:
# one emitted spectral packet
# -> broadcasts through field
# -> many candidate receivers listen
# -> only matching receiver filters absorb
# -> fanout/collision is controlled
# -> weak / wrong / ambiguous branches are rejected
# ============================================================

np.random.seed(33)

STATE = "D33_SPECTRAL_CASCADE_BRANCHING"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
branch_fanout_default = 2

receiver_target_sequence = [
    "D4_O5",
    "D1_O2",
    "D4_O5",
    "D3_O6",
    "D0_O1",
    "D5_O4",
    "D2_O3",
    "D4_O5",
]

receiver_initial_sequence = ["EMPTY"] * len(receiver_target_sequence)


def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE", "NOISE"]:
        return None, None

    if not isinstance(key, str) or "_" not in key:
        return None, None

    d_part, o_part = key.split("_", 1)

    if not d_part.startswith("D") or not o_part.startswith("O"):
        return None, None

    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))
    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def wave_token(key, prefix="BRANCH_PHOTON"):
    d, o = parse_key(key)
    if d is None:
        return "NONE"
    phase = round(d / directions, 3)
    frequency = o + 1
    return f"{prefix}[D{d}|O{o}|F{frequency}|P{phase}]"


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0,
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.18 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE", "NOISE"]:
        d, o = parse_key(target_key)
        if d is not None:
            packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE", "NOISE"]:
        d2, o2 = parse_key(ambiguous_key)
        if d2 is not None:
            packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score
    accepted = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "accepted": accepted,
    }


def key_color(key, strength=1.0):
    if key in [None, "EMPTY", "NONE"]:
        return "black"

    if key == "WHITE":
        return "white"

    if key == "NOISE":
        return "gray"

    d, o = parse_key(key)
    if d is None:
        return "black"

    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)
    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def draw_arrow(ax, start, end, active=True, rejected=False, blocked=False):
    if blocked:
        color = "red"
    elif rejected:
        color = "gray"
    else:
        color = "orange" if active else "lightgray"

    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=12,
        linewidth=1.8,
        color=color,
        alpha=alpha,
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.15
    gap = 0.03

    ax.text(x0, y0 + 1.38, title, fontsize=6, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.3 if is_selected else 0.25

            ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))


def draw_memory_slots(ax, x0, y0, seq, title, active_slots=None, blocked_slots=None):
    if active_slots is None:
        active_slots = []
    if blocked_slots is None:
        blocked_slots = []

    cell = 0.19
    gap = 0.04

    ax.text(x0, y0 + 1.9, title, fontsize=6, ha="left", va="bottom", color="black")

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)

        if i in active_slots:
            edge = "lime"
            lw = 1.5
        elif i in blocked_slots:
            edge = "red"
            lw = 1.5
        else:
            edge = "black"
            lw = 0.4

        ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))
        ax.text(x + 0.28, y + cell / 2, str(i), fontsize=5.5, ha="left", va="center", color="black")


def draw_receivers(ax, candidate_slots, absorbed_slots, blocked_slots, decoded_key, memory_state):
    receiver_positions = {
        0: (5.1, 4.1),
        1: (6.2, 4.1),
        2: (7.3, 4.1),
        3: (5.1, 3.0),
        4: (6.2, 3.0),
        5: (7.3, 3.0),
        6: (5.65, 1.9),
        7: (6.75, 1.9),
    }

    for slot, (x, y) in receiver_positions.items():
        is_candidate = slot in candidate_slots
        is_absorbed = slot in absorbed_slots
        is_blocked = slot in blocked_slots

        current_key = memory_state[slot]
        filter_key = receiver_target_sequence[slot]

        if is_absorbed:
            face = key_color(decoded_key)
            edge = "lime"
            lw = 2.0
        elif is_blocked:
            face = key_color(filter_key)
            edge = "red"
            lw = 2.0
        elif is_candidate:
            face = key_color(filter_key)
            edge = "orange"
            lw = 1.5
        else:
            face = key_color(current_key)
            edge = "black"
            lw = 0.7

        ax.add_patch(Circle((x, y), 0.32, facecolor=face, edgecolor=edge, linewidth=lw))
        ax.text(x, y - 0.52, f"S{slot}", fontsize=6, ha="center", color="black")
        ax.text(x, y + 0.45, filter_key, fontsize=5.5, ha="center", color="black")


def draw_snapshot(ax, event_row, packet, memory_state):
    decoded_key = event_row["DECODED_KEY"]
    source_key = event_row["SOURCE_KEY"]
    emitted_key = event_row["EMITTED_KEY"]

    candidate_slots = event_row["CANDIDATE_SLOTS_LIST"]
    absorbed_slots = event_row["ABSORBED_SLOTS_LIST"]
    blocked_slots = event_row["BLOCKED_SLOTS_LIST"]

    ax.set_facecolor("white")
    ax.set_xlim(0, 12.5)
    ax.set_ylim(0.55, 5.2)
    ax.set_aspect("equal")
    ax.axis("off")

    source_color = key_color(source_key)
    emitted_color = key_color(emitted_key)
    decoded_color = key_color(decoded_key)

    ax.add_patch(Circle((0.8, 3.2), 0.35, facecolor=source_color, edgecolor="black", linewidth=1.0))
    ax.text(0.8, 2.65, f"SRC\n{source_key}", fontsize=6, ha="center", color="black")

    ax.add_patch(Circle((1.7, 3.2), 0.25, facecolor=emitted_color, edgecolor="black", linewidth=1.0))
    ax.text(1.7, 2.65, "EMIT", fontsize=6, ha="center", color="black")

    ax.add_patch(Rectangle((2.45, 3.0), 0.38, 0.38, facecolor="limegreen", edgecolor="black", linewidth=1.0))
    ax.text(2.64, 2.65, "FIELD", fontsize=6, ha="center", color="black")

    ax.add_patch(Circle(
        (3.65, 3.2),
        0.35,
        facecolor=decoded_color,
        edgecolor="lime" if event_row["PACKET_ACCEPTED"] else "red",
        linewidth=2.0,
    ))
    ax.text(3.65, 2.55, f"DECODE\n{decoded_key}", fontsize=6, ha="center", color="black")

    draw_arrow(ax, (1.15, 3.2), (1.45, 3.2), active=True)
    draw_arrow(ax, (1.95, 3.2), (2.45, 3.2), active=True)
    draw_arrow(ax, (2.83, 3.2), (3.3, 3.2), active=True, rejected=not event_row["PACKET_ACCEPTED"])

    draw_receivers(ax, candidate_slots, absorbed_slots, blocked_slots, decoded_key, memory_state)

    receiver_positions = {
        0: (5.1, 4.1),
        1: (6.2, 4.1),
        2: (7.3, 4.1),
        3: (5.1, 3.0),
        4: (6.2, 3.0),
        5: (7.3, 3.0),
        6: (5.65, 1.9),
        7: (6.75, 1.9),
    }

    for slot in candidate_slots:
        pos = receiver_positions[slot]
        draw_arrow(
            ax,
            (4.0, 3.2),
            (pos[0] - 0.32, pos[1]),
            active=True,
            rejected=(slot not in absorbed_slots),
            blocked=(slot in blocked_slots),
        )

    draw_packet_grid(ax, 8.35, 2.55, packet, "42-channel branch packet", selected_key=decoded_key)
    draw_memory_slots(
        ax,
        10.65,
        1.55,
        memory_state,
        "branch memory",
        active_slots=absorbed_slots,
        blocked_slots=blocked_slots,
    )

    ax.text(
        0.55,
        0.78,
        f"case={event_row['CASE']} decoded={decoded_key} score={event_row['BEST_SCORE']} "
        f"margin={event_row['MARGIN']} fanout={event_row['FANOUT_LIMIT']} "
        f"absorbed={event_row['ABSORBED_SLOTS']} blocked={event_row['BLOCKED_SLOTS']}\n"
        f"{event_row['OUTPUT_TOKEN']}",
        fontsize=6,
        ha="left",
        color="black",
    )

    ax.set_title(
        f"D33 snapshot t={event_row['T']} — {event_row['CASE']} / {event_row['ACTION']}",
        fontsize=9,
        color="black",
    )


cases = [
    {
        "case": "BRANCH_FANOUT_0_2",
        "source_key": "D4_O5",
        "emitted_key": "D4_O5",
        "candidate_slots": [0, 2, 7],
        "fanout_limit": 2,
        "packet": spectral_packet("D4_O5", strength=0.91),
    },
    {
        "case": "SINGLE_BRANCH_1",
        "source_key": "D1_O2",
        "emitted_key": "D1_O2",
        "candidate_slots": [1],
        "fanout_limit": 1,
        "packet": spectral_packet("D1_O2", strength=0.92),
    },
    {
        "case": "WHITE_TARGET_BRANCH_3",
        "source_key": "WHITE",
        "emitted_key": "D3_O6",
        "candidate_slots": [3],
        "fanout_limit": 1,
        "packet": spectral_packet("D3_O6", strength=0.90, broad=True),
    },
    {
        "case": "LOW_OCTAVE_BRANCH_4",
        "source_key": "D0_O1",
        "emitted_key": "D0_O1",
        "candidate_slots": [4],
        "fanout_limit": 1,
        "packet": spectral_packet("D0_O1", strength=0.89),
    },
    {
        "case": "HIGH_DIRECTION_BRANCH_5",
        "source_key": "D5_O4",
        "emitted_key": "D5_O4",
        "candidate_slots": [5],
        "fanout_limit": 1,
        "packet": spectral_packet("D5_O4", strength=0.90),
    },
    {
        "case": "MID_BRANCH_6",
        "source_key": "D2_O3",
        "emitted_key": "D2_O3",
        "candidate_slots": [6],
        "fanout_limit": 1,
        "packet": spectral_packet("D2_O3", strength=0.93),
    },
    {
        "case": "LOOPBACK_FILL_7",
        "source_key": "D4_O5",
        "emitted_key": "D4_O5",
        "candidate_slots": [7],
        "fanout_limit": 1,
        "packet": spectral_packet("D4_O5", strength=0.88),
    },
    {
        "case": "WRONG_DIRECTION_REJECT",
        "source_key": "D0_O2",
        "emitted_key": "D0_O2",
        "candidate_slots": [1, 4],
        "fanout_limit": 1,
        "packet": spectral_packet("D0_O2", strength=0.90),
    },
    {
        "case": "WRONG_OCTAVE_REJECT",
        "source_key": "D4_O1",
        "emitted_key": "D4_O1",
        "candidate_slots": [0, 2, 7],
        "fanout_limit": 1,
        "packet": spectral_packet("D4_O1", strength=0.88),
    },
    {
        "case": "WEAK_BRANCH_REJECT",
        "source_key": "D5_O4",
        "emitted_key": "D5_O4",
        "candidate_slots": [5],
        "fanout_limit": 1,
        "packet": spectral_packet("D5_O4", strength=0.39),
    },
    {
        "case": "AMBIGUOUS_BRANCH_REJECT",
        "source_key": "D2_O3",
        "emitted_key": "D2_O3",
        "candidate_slots": [6],
        "fanout_limit": 1,
        "packet": spectral_packet("D2_O3", strength=0.86, ambiguous_key="D4_O5", ambiguous_strength=0.81),
    },
]

receiver_memory = receiver_initial_sequence.copy()

event_rows = []
receiver_rows = []
memory_rows = []
wave_rows = []
snapshots = []

accepted_packets = 0
rejected_packets = 0
absorbed_receivers = 0
blocked_by_fanout = 0

for event_index, item in enumerate(cases):
    decoded = decode_packet(item["packet"])
    decoded_key = decoded["best_key"]
    packet_accepted = bool(decoded["accepted"])

    candidate_slots = list(item["candidate_slots"])

    matching_slots = [
        slot
        for slot in candidate_slots
        if packet_accepted and receiver_target_sequence[slot] == decoded_key
    ]

    fanout_limit = int(item.get("fanout_limit", branch_fanout_default))

    selected_slots = matching_slots[:fanout_limit]
    blocked_slots = matching_slots[fanout_limit:]

    if packet_accepted:
        accepted_packets += 1
    else:
        rejected_packets += 1

    blocked_by_fanout += len(blocked_slots)

    before_memory = receiver_memory.copy()

    for slot in selected_slots:
        receiver_memory[slot] = decoded_key
        absorbed_receivers += 1

    after_memory = receiver_memory.copy()

    if selected_slots:
        action = "APPLY_BRANCH_ABSORB"
    elif blocked_slots:
        action = "REJECT_FANOUT_COLLISION"
    elif packet_accepted:
        action = "REJECT_FILTER_MISMATCH"
    else:
        action = "REJECT_WEAK_OR_AMBIGUOUS"

    event_row = {
        "STATE": STATE,
        "T": event_index * 120,
        "CASE": item["case"],
        "SOURCE_KEY": item["source_key"],
        "EMITTED_KEY": item["emitted_key"],
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "PACKET_ACCEPTED": int(packet_accepted),
        "CANDIDATE_SLOTS": ",".join(str(x) for x in candidate_slots),
        "MATCHING_SLOTS": ",".join(str(x) for x in matching_slots) if matching_slots else "none",
        "ABSORBED_SLOTS": ",".join(str(x) for x in selected_slots) if selected_slots else "none",
        "BLOCKED_SLOTS": ",".join(str(x) for x in blocked_slots) if blocked_slots else "none",
        "FANOUT_LIMIT": fanout_limit,
        "ACTION": action,
        "OUTPUT_TOKEN": wave_token(decoded_key if packet_accepted else "NONE"),
        "CANDIDATE_SLOTS_LIST": candidate_slots,
        "ABSORBED_SLOTS_LIST": selected_slots,
        "BLOCKED_SLOTS_LIST": blocked_slots,
    }

    event_rows.append(event_row)

    for slot in candidate_slots:
        filter_key = receiver_target_sequence[slot]
        before_key = before_memory[slot]
        after_key = after_memory[slot]

        filter_match = packet_accepted and decoded_key == filter_key
        absorbed = slot in selected_slots
        blocked = slot in blocked_slots

        if absorbed:
            slot_action = "ABSORB_AND_WRITE"
        elif blocked:
            slot_action = "BLOCKED_BY_FANOUT"
        elif not packet_accepted:
            slot_action = "REJECT_WEAK_OR_AMBIGUOUS"
        else:
            slot_action = "REJECT_FILTER_MISMATCH"

        receiver_rows.append({
            "STATE": STATE,
            "CASE": item["case"],
            "RECEIVER_SLOT": slot,
            "FILTER_KEY": filter_key,
            "BEFORE_KEY": before_key,
            "DECODED_KEY": decoded_key,
            "AFTER_KEY": after_key,
            "PACKET_ACCEPTED": int(packet_accepted),
            "FILTER_MATCH": int(filter_match),
            "ABSORBED": int(absorbed),
            "BLOCKED_BY_FANOUT": int(blocked),
            "ACTION": slot_action,
            "MATCH_IF_ABSORBED": int(after_key == filter_key) if absorbed else "not_written",
            "HELD_IF_REJECTED": int(after_key == before_key) if not absorbed else 0,
        })

    wave_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "SOURCE_KEY": item["source_key"],
        "EMITTED_KEY": item["emitted_key"],
        "DECODED_KEY": decoded_key,
        "PACKET_ACCEPTED": int(packet_accepted),
        "ABSORBED_SLOTS": ",".join(str(x) for x in selected_slots) if selected_slots else "none",
        "BLOCKED_SLOTS": ",".join(str(x) for x in blocked_slots) if blocked_slots else "none",
        "OUTPUT_TOKEN": wave_token(decoded_key if packet_accepted else "NONE"),
    })

    snapshots.append((event_row, item["packet"], after_memory.copy()))

df_events = pd.DataFrame(event_rows).drop(columns=[
    "CANDIDATE_SLOTS_LIST",
    "ABSORBED_SLOTS_LIST",
    "BLOCKED_SLOTS_LIST",
])
df_receivers = pd.DataFrame(receiver_rows)
df_wave = pd.DataFrame(wave_rows)

for slot, key in enumerate(receiver_memory):
    memory_rows.append({
        "STATE": STATE,
        "MEMORY_SLOT": slot,
        "TARGET_KEY": receiver_target_sequence[slot],
        "FINAL_KEY": key,
        "MATCH_TARGET": int(key == receiver_target_sequence[slot]),
    })

df_memory = pd.DataFrame(memory_rows)

branch_accuracy = float(np.mean(np.array(receiver_memory) == np.array(receiver_target_sequence)))

receiver_safety_accuracy = float(np.mean([
    1 if (
        (row["ABSORBED"] == 1 and row["AFTER_KEY"] == row["FILTER_KEY"]) or
        (row["ABSORBED"] == 0 and row["HELD_IF_REJECTED"] == 1)
    ) else 0
    for row in receiver_rows
]))

result = (
    "SPECTRAL_CASCADE_BRANCHING_LOCKED"
    if branch_accuracy == 1.0 and receiver_safety_accuracy == 1.0
    else "SPECTRAL_CASCADE_BRANCHING_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": len(receiver_memory),
    "TARGET_SEQUENCE": " ".join(receiver_target_sequence),
    "FINAL_SEQUENCE": " ".join(receiver_memory),
    "CASES_TESTED": len(cases),
    "ACCEPTED_PACKETS": accepted_packets,
    "REJECTED_PACKETS": rejected_packets,
    "ABSORBED_RECEIVERS": absorbed_receivers,
    "BLOCKED_BY_FANOUT": blocked_by_fanout,
    "BRANCH_ACCURACY": round(branch_accuracy, 3),
    "RECEIVER_SAFETY_ACCURACY": round(receiver_safety_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "DEFAULT_FANOUT": branch_fanout_default,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8, 3.15 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {len(receiver_memory)}")
print(f"target sequence:              {' '.join(receiver_target_sequence)}")
print(f"final sequence:               {' '.join(receiver_memory)}")
print(f"cases tested:                 {len(cases)}")
print(f"accepted packets:             {accepted_packets}")
print(f"rejected packets:             {rejected_packets}")
print(f"absorbed receivers:           {absorbed_receivers}")
print(f"blocked by fanout:            {blocked_by_fanout}")
print(f"branch accuracy:              {round(branch_accuracy, 3)}")
print(f"receiver safety accuracy:     {round(receiver_safety_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"default fanout:               {branch_fanout_default}")

print("\nBranch event table:")
display(df_events)

print("\nReceiver decision table:")
display(df_receivers)

print("\nFinal branch memory table:")
display(df_memory)

print("\nBranch photon table:")
display(df_wave)

print("\nSummary:")
display(df_summary)
```
